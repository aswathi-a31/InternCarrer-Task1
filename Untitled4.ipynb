```python
df=df.dropna()
```

```python
# Get unique values in the "Categories" column
unique_categories = df['Categories'].unique()

# Print the unique categories
print("Unique Categories:")
for category in unique_categories:
    print(category)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)

# Mean imputation for numeric columns
df['Suscribers'].fillna(df['Suscribers'].mean(), inplace=True)

# Repeat this for other numeric columns with missing values

# Check if missing values have been handled
missing_values_after_imputation = df.isnull().sum()
print("\nMissing Values After Imputation:")
print(missing_values_after_imputation)
```

```python
# Data Analysis
# Find the top channels with the most subscribers
top_subscribers = df.sort_values(by='Suscribers', ascending=False).head(10)
print("\nTop Channels by Subscribers:")
print(top_subscribers)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_subscribers['Username'], top_subscribers['Suscribers'])
plt.xlabel("Channel Username")
plt.ylabel("Subscribers")
plt.title("Top Channels by Subscribers")
plt.xticks(rotation=45)
plt.show()
```

```python
# List of categorical column names you want to explore
categorical_columns = ['Categories', 'Country']

# Explore each categorical column
for column in categorical_columns:
    unique_values = df[column].unique()
    num_unique_values = len(unique_values)

    # Count the frequency of each unique value in the column
    value_counts = df[column].value_counts()

    print(f"Exploring column: {column}")
    print(f"Number of unique values: {num_unique_values}")

    # Display unique values and their frequencies
    print("Unique Values and Their Frequencies:")
    for value, count in value_counts.items():
        print(f"{value}: {count} ({count / len(df) * 100:.2f}%)")

    # Separate the output of different columns
    print("\n")
```

```python
# List of categorical column names you want to explore
categorical_columns = ['Categories', 'Country']  # Add more columns as needed

# Create bar charts for each categorical column
for column in categorical_columns:
    value_counts = df[column].value_counts()

    # Plot a bar chart
    plt.figure(figsize=(10, 6))
    value_counts.plot(kind='bar', color='skyblue')
    plt.title(f'Frequency Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=75)  # Rotate x-axis labels for readability
    plt.show()
```

```python
# Create a histogram for the 'Suscribers' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Suscribers'], bins=20, kde=True)
plt.title('Distribution of Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a box plot to visualize 'Likes' by 'Categories'
plt.figure(figsize=(10, 8))
sns.boxplot(x='Categories', y='Likes', data=df)
plt.title('Likes by Category')
plt.xlabel('Category')
plt.ylabel('Likes')
plt.xticks(rotation=80)
plt.show()
```

```python
# Create a count plot for the 'Country' column
plt.figure(figsize=(10, 6))
sns.countplot(x='Country', data=df, palette='viridis')
plt.title('Count of Channels by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a pair plot to explore relationships between numerical columns
sns.pairplot(df[['Suscribers', 'Visits', 'Likes', 'Comments']])
plt.show()
```

```python
# Group by 'Categories' and calculate the average subscribers and likes within each category
grouped_by_category = df.groupby('Categories').agg({
    'Suscribers': 'mean',
    'Likes': 'mean'
}).reset_index()

print("Average Subscribers and Likes by Category:")
print(grouped_by_category)
```

```python
# Group by 'Country' and calculate the total subscribers and likes within each country
grouped_by_country = df.groupby('Country').agg({
    'Suscribers': 'sum',
    'Likes': 'sum'
}).reset_index()

print("Total Subscribers and Likes by Country:")
print(grouped_by_country)
```

```python
# Create a bar chart to visualize the average subscribers and likes by category
plt.figure(figsize=(12, 6))
plt.bar(grouped_by_category['Categories'], grouped_by_category['Suscribers'], label='Average Subscribers', alpha=0.7)
plt.bar(grouped_by_category['Categories'], grouped_by_category['Likes'], label='Average Likes', alpha=0.7)
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Average Subscribers and Likes by Category')
plt.xticks(rotation=80)
plt.legend()
plt.show()
```

```python
# Create a bar chart to visualize the total subscribers and likes by country
plt.figure(figsize=(12, 6))
plt.bar(grouped_by_country['Country'], grouped_by_country['Suscribers'], label='Total Subscribers', alpha=0.7)
plt.bar(grouped_by_country['Country'], grouped_by_country['Likes'], label='Total Likes', alpha=0.7)
plt.xlabel('Country')
plt.ylabel('Counts')
plt.title('Total Subscribers and Likes by Country')
plt.xticks(rotation=65)
plt.legend()
plt.show()
```

```python
# Calculate the distribution of categories
category_counts = df['Categories'].value_counts()

# Create a pie chart to visualize the distribution of categories
plt.figure(figsize=(15, 30))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Categories')
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between subscribers and likes
plt.figure(figsize=(10, 5))
plt.scatter(df['Suscribers'], df['Likes'], alpha=0.5)
plt.xlabel('Subscribers')
plt.ylabel('Likes')
plt.title('Scatter Plot of Subscribers vs. Likes')
plt.show()
```

```python
average_subscribers = df['Suscribers'].mean()
average_visits = df['Visits'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()
```

```python
above_average_streamers = df[(df['Suscribers'] > average_subscribers) &
                             (df['Visits'] > average_visits) &
                             (df['Likes'] > average_likes) &
                             (df['Comments'] > average_comments)]
```

```python
top_performing_streamers = above_average_streamers.sort_values(by='Suscribers', ascending=False)
```

```python
print("Top-Performing Content Creators:")
print(top_performing_streamers[['Username', 'Categories', 'Suscribers', 'Visits', 'Likes', 'Comments']])
```
